```python
import numpy as np
import os
# Define directories
directoryDataPointingInfo = "/mnt/c/Users/calsifer/Desktop/RGC/"
directoryDataImages = "/mnt/c/Users/calsifer/Desktop/RGC/data/"

# Define the file name containing pointing information
fileNamePointingInfo = "matched_targets_test.csv"
pathPointingInfo = directoryDataPointingInfo + fileNamePointingInfo

# Load pointing names from the CSV file (assuming the pointing names are in the 5th column)
pointingNamesAll = np.loadtxt(pathPointingInfo, usecols=5, delimiter=",", dtype=str)

# Loop through each pointing name
for pointingName in pointingNamesAll:
    # Construct the path for the image
    pathImage = os.path.join(directoryDataImages, f"{pointingName}-mosaic.fits")

    # Print progress
    print(f"Download data for pointing {pointingName}...")

    # Check if the file already exists
    if os.path.exists(pathImage):
        print(f"The image for pointing {pointingName} already exists at '{pathImage}'.")
    else:
        # Download the file using wget, and rename it with the pointing name in the file
        os.system(f"wget -nc --user XXX --password YYY https://lofar-surveys.org/public/DR2/mosaics/{pointingName}/low-mosaic-blanked.fits -O '{pathImage}'")
        print(f"Downloaded data for pointing {pointingName} as {pathImage}.")
```
